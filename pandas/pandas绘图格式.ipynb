# pandas绘图格式
**通过设置样式参数，创建符合个人喜好的图表风格**

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

## 1、通用样式参数

```python
gs = pd.Series(np.random.randint(1,10,10))
# 线条颜色
# style in [r--,g--,b--,y--,k--]
gs.plot(style='y--')
```

## 2、控制说明

```python
legend_df = pd.DataFrame(np.random.randint(1,10,(10,2)),columns=list('ab'))
# 针对legend 是针对dataframe的列名
legend_df.plot(legend=False)
```

## 3、时间序列

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index, columns=list('ABCD'))
df = df.cumsum()
```

```python
# 抑制图
df.A.plot(x_compat=True)
```

```python
# 设置多个抑制图
with pd.plotting.plot_params.use('x_compat', True):
    df.A.plot(color='r')
    df.B.plot(color='g')
    df.C.plot(color='b')
```

## 4、设置子图

```python
df.plot(subplots=True,figsize=(6,6))
```

```python
# 设置是否共享x轴 sharex=False
df.plot(subplots=True,figsize=(6,6),sharex=False)
```

```python
# 设置子图布局
# layout=(x,y) x为行，y为列
df.plot(subplots=True,figsize=(16,6),sharex=False,layout=(2,3))
```

```python
# 设置子图布局
fig, axes = plt.subplots(nrows=2, ncols=2)
df['A'].plot(ax=axes[0,0]); axes[0,0].set_title('A');
df['B'].plot(ax=axes[0,1]); axes[0,1].set_title('B');
df['C'].plot(ax=axes[1,0]); axes[1,0].set_title('C');
df['D'].plot(ax=axes[1,1]); axes[1,1].set_title('D');
```
