# pandas绘图工具

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

## 1、散点矩阵图

```python
from pandas.plotting import scatter_matrix
sm = pd.DataFrame(np.random.randn(100,4),columns=list('abcd'))
# diagonal : {'hist', 'kde'}
scatter_matrix(sm,diagonal='kde')
```

## 2、密度图

```python
dp = pd.DataFrame(np.random.randn(100,4),columns=list('abcd'))
# 根据某一列的值画图
dp['a'].plot.kde()

# 根据所有列的值画图
dp.plot.kde()
```

## 3、RadViz图
**RadViz是一种可视化多变量数据的方式。它基于简单的弹簧张力最小化算法。基本上你在飞机上设置了一堆点。在我们的例子中，它们在单位圆上的距离相等。每个点代表一个属性。然后假设数据集中的每个样本都通过弹簧与这些点中的每个点相连，其刚度与该属性的数值成比例（它们被归一化为单位间隔）。在我们的样品稳定在哪里（作用于我们样品的力处于平衡状态）的平面上的点是绘制表示我们样品的点的位置。根据样本属于哪个类别，它将以不同的颜色着色。**

```python
from pandas.plotting import radviz
rv = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/master/pandas/tests/data/iris.csv")
radviz(rv,'Name')
```

## 4、平行坐标
**平行坐标允许查看数据中的聚类，并以可视方式估计其他统计数据。使用平行坐标点表示为连接的线段。每条垂直线代表一个属性。一组连接的线段表示一个数据点。倾向于聚类的点将显得更加接近**

```python
from pandas.plotting import parallel_coordinates
# pz = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/master/pandas/tests/data/iris.csv")
parallel_coordinates(rv,'Name')
```
