# pandas绘图

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

## 1、基本绘制
**series和Dataframe的plot()方法**

```python
# series
ts = pd.Series(range(10))
ts.plot()
```

```python
# dataframe
df = pd.DataFrame(np.random.randint(1,100,(10,2)),columns=list('ab'))
plt.figure()
df.plot()
```

```python
# dataframe 设置列与画图轴的关系
df1 = df.cumsum()
df1['c'] = pd.Series(range(len(df)))
df1.plot(x='a',y='b')
# df1.plot(x='a',y='c')
```

### 2.2、条形图: bar/barh
* bar 用于竖向
* barh 用于横向

```python
bar_df = pd.DataFrame(np.random.randint(1,100,(10,4)),columns=list('abcd'))
bar_df.plot.bar()
```

```python
bar_df.plot.bar()
# 设置x轴辅助线
plt.axhline(0, color='green')
```

```python
bar_df
```

```python
# 类别重叠
bar_df.plot.bar(stacked=True)
```

### 2.3、直方图
* hist

```python
hist_df = pd.DataFrame(np.random.randn(100,4),columns=list('abcd'))
hist_df
```

```python
hist_df.plot.hist()
```

```python
# 设置透明图
# alpha => [0,1]之间
hist_df.plot.hist(alpha=0.4)
```

```python
# 设置重叠
hist_df.plot.hist(stacked=True,bins=20)
```

```python
# 设置旋转
hist_df.plot.hist(orientation='horizontal', cumulative=True)
```

### 2.4、盒图
* box
* boxplot

```python
box_df = pd.DataFrame(np.random.randint(1,10,(10,2)),columns=list('ab'))
display(box_df.max(),box_df.min(),box_df)
```

```python
box_df.plot.box()
```

```python
# 旋转
box_df.plot.box(vert=False)
```

```python
# dataframe本身封装好的函数boxplot()
# 从结果图中可以看出，只是界面有横向标线标明
box_df.boxplot()
```

```python
# 分组操作
box_df['X'] = pd.Series(['A','A','B','A','B','B','B','A','A','B'])
box_df.boxplot(by='X')
```

### 2.5、区域图
**区域图是默认堆叠的。要生成堆积面积图，每列必须全部为正值或全为负值。**

```python
area_df = pd.DataFrame(np.random.randint(1,10,(10,2)),columns=list('ab'))
area_df
```

```python
# 默认情况是先画第一列的区域，然后再根据第一列的每个x轴所对应的y轴的值，为起点画第二列每个x轴所对应的y轴的值
area_df.plot.area()
```

```python
area_df.plot.area(stacked=False)
```

### 2.6、散点图
**散点图需要x和y轴的数字列**

```python
scatter_df = pd.DataFrame(np.random.randint(1,100,(20,3)),columns=list('abc'))
scatter_df
```

```python
# 指定散点图的x/y轴所对应哪些列
scatter_df.plot.scatter(x='a',y='b')
```

```python
# 设置多组散点图
ab_ax = scatter_df.plot.scatter(x='a', y='b', color='DarkBlue', label='aAndb');
# bc_ax = scatter_df.plot.scatter(x='b', y='c', color='red', label='bAndc');
scatter_df.plot.scatter(x='a', y='c', color='DarkGreen', label='aAndc', ax=ab_ax);
```

```python
# 设置泡泡大小,一致
scatter_df.plot.scatter(x='a',y='b',s=50)
```

```python
# 设置泡泡大小
scatter_df.plot.scatter(x='a',y='b',s=scatter_df['c'])
```

### 2.7、饼图

```python
pie_df = pd.DataFrame(np.random.randint(1,10,(10,4)),columns=list('abcd'))
# series = pd.Series(3 * np.random.rand(4), index=['a', 'b', 'c', 'd'], name='series')
# series
```

```python
# 根据a列的数值画饼图
pie_df['a'].plot.pie()
```

```python
# 根据多列的数值画出饼图
pie_df.plot.pie(subplots=True, figsize=(16, 4))
```

```python
# 根据a列的数值画出每个数值在饼图的占比
pie_df['a'].plot.pie(autopct='%.2f', fontsize=20, figsize=(6, 6))
```
