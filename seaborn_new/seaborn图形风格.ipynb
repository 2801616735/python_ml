# seaborn图形风格
**要控制样式，请使用axes_style()和set_style()功能。要缩放图，请使用plotting_context()和set_context()函数。**

```python
import  numpy  as  np 
# import  matplotlib  as  mpl 
import  matplotlib.pyplot  as  plt 
import  seaborn  as  sns
%matplotlib inline
np.random.seed(sum(map(ord, "aesthetics")))
```

```python
# 定义一个简单函数来绘制一些偏移正弦波
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
sinplot()
```

## 1、切换默认seaborn样式

```python
sns.set()
sinplot()
```

## 2、seaborn绘制样式

```python
# darkgrid, whitegrid, dark, white, and ticks
sns.set_style("whitegrid")
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.boxplot(data=data)
```

## 3、去除轴脊椎

```python
# sns.set_style("ticks")
sinplot()
sns.despine()
```

## 4、临时seaborn绘制样式

```python
with sns.axes_style("darkgrid"):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
```

## 5、重新seaborn绘制样式

```python
# 查看可使用的样式
sns.axes_style()
```

```python
sns.set_style('darkgrid',{'axes.facecolor':'.9'})
sinplot()
```

## 6、绘制元素缩放

```python
# 查看可使用的控制函数
sns.plotting_context()
```

```python
# 控制线条大小
sns.set_context(rc={'lines.linewidth':0.5})
sinplot()
```

```python
# paper, notebook, talk, poster 线条由小到大
sns.set_context('notebook')
sinplot()
```
