# matplotlib子图绘制

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline

# 载入数据
data = pd.read_csv("../file/date.csv")
```

## 1、子图划分

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
plt.show()
```

## 2、子图绘制

```python
ax1.plot(np.random.randint(1,10,(5)),np.arange(5))
ax2.plot(np.arange(10)*3,np.arange(10))
plt.show()
```

## 3、在同一个画板绘制多条

```python
# 第一条
plt.plot(data['date'][0:10],data['ceil'][0:10],'r--')
# 第二条
plt.plot(data['date'][0:10],data['ceil'][11:21],'g--')
# x轴日期显示优化
plt.xticks(rotation=45)

plt.show()
```

## 4、增加label提示

```python
color = ['blue','green']
for i in range(2):
    s = 10*i
    e = 10*(i+1)
    plt.plot(data['date'][0:10],data['ceil'][s:e],color[i],label=i)

# plt.legend(loc=0)
plt.legend(loc=1)
# plt.legend(loc=2)

# x轴日期显示优化
plt.xticks(rotation=45)
plt.xlabel('date')
plt.ylabel('value')
plt.title('date-value')
plt.show()
    
```
