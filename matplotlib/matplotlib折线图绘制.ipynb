# matplotlib折线图绘制

```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

## 1、折线图绘制
**需要确定x轴和y轴的值**

```python
# 如无传递参数，则显示空白
plt.plot()
plt.show()
```

```python
file_name = "../file/date.csv"
```

```python
# 获取数据
data = pd.read_csv(file_name)

# 把数据代入x轴和y轴
plt.plot(data['date'][0:10],data['ceil'][0:10])
plt.show()
```

```python
# 把数据代入x轴和y轴
plt.plot(data['date'][0:10],data['ceil'][0:10])

# 优化x轴的label
plt.xticks(rotation=45)

plt.show()
```

```python
# 把数据代入x轴和y轴
plt.plot(data['date'][0:10],data['ceil'][0:10])

# 优化x轴的label
plt.xticks(rotation=45)
plt.xlabel('date')
plt.ylabel('value')
plt.title('date and value')

plt.show()
```
