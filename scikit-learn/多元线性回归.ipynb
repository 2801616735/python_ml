# 多元线性回归

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
%matplotlib inline
# 显示中文
from pylab import mpl 
mpl.rcParams['font.sans-serif'] = ['SimHei']

def runplt():
    plt.figure()
    plt.title('匹萨价格与直径数据')
    plt.xlabel('直径（英寸）')
    plt.ylabel('价格（美元）')
    # 定义好x轴和y轴的刻度
    plt.axis([0, 25, 0, 25])
    plt.grid(True)
    return plt
X_train = [[6], [8], [10], [14], [18]]
y_train = [[7], [9], [13], [17.5], [18]]
X_test = [[4],[6], [8], [11], [16],[18],[21]]
y_test = [[6],[8], [12], [15], [18],[20],[23]]
regressor = LinearRegression()
regressor.fit(X_train, y_train)
yy = regressor.predict(X_test)
plt = runplt()
plt.plot(X_train, y_train, 'k.')
plt.plot(X_test, yy)

quadratic_featurizer = PolynomialFeatures(degree=2)
X_train_quadratic = quadratic_featurizer.fit_transform(X_train)
X_test_quadratic = quadratic_featurizer.transform(X_test)
regressor_quadratic = LinearRegression()
regressor_quadratic.fit(X_train_quadratic, y_train)
plt.plot(X_test, regressor_quadratic.predict(X_test_quadratic), 'r-')
plt.show()
# print(X_train)
# print(X_train_quadratic)
# print(X_test)
# print(X_test_quadratic)
print('一元线性回归 r-squared', regressor.score(X_test, y_test))
print('二次回归 r-squared', regressor_quadratic.score(X_test_quadratic, regressor_quadratic.predict(X_test_quadratic)))
```
