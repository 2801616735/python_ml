# 一元线性回归

```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
%matplotlib inline
# 显示中文
from pylab import mpl 
mpl.rcParams['font.sans-serif'] = ['SimHei']
```

## 1、训练集数据

```python
# 训练样本	直径（英寸）	价格（美元）
# 1	6	7
# 2	8	9
# 3	10	13
# 4	14	17.5
# 5	18	18
def runplt():
    plt.figure()
    plt.title('匹萨价格与直径数据')
    plt.xlabel('直径（英寸）')
    plt.ylabel('价格（美元）')
    # 定义好x轴和y轴的刻度
    plt.axis([0, 25, 0, 25])
    plt.grid(True)
    return plt

plt = runplt()
X = [[6], [8], [10], [14], [18]]
y = [[7], [9], [13], [17.5], [18]]
plt.plot(X, y, 'k.')
plt.show()
```

## 2、使用线性模型

```python
from sklearn.linear_model import LinearRegression
# 创建模型并训练
model = LinearRegression()
model.fit(X,y)

# 进行预测操作
print("预测14英寸的披萨价格是:%.2f"% model.predict([[14]])[0,0] )
```

### 2.1、显示预测结果

```python
plt = runplt()
plt.plot(X, y, 'k.')
plt.plot(X, model.predict(X), 'r-')
plt.show()
```

### 2.2、评估

```python
from sklearn import metrics
# metrics.r2_score(y,model.predict(X))
model.score(X,model.predict(X))
```
