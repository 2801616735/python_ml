# seaborn主题风格设置

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1、matplotlib绘制

```python
def sinplot(flip=1):
    x = np.linspace(0,10,100)
    for i in range(1,5):
        plt.plot(x,np.sin(x+i*0.5)*(5-i)*flip)

sinplot()
```

## 2、seaborn主题设置

```python
sns.set()
sinplot()
```

## 2.1、darkgrid

```python
# darkgrid, whitegrid, dark, white, ticks
sns.set_style('darkgrid')
data = np.random.normal(size=(20,5)) + np.arange(5)/2
sns.boxplot(data=data)
```

## 2.2、whitegrid

```python
sns.set_style('whitegrid')
sinplot()
```

## 2.3、dark

```python
sns.set_style('dark')
sinplot()
```

## 2.4、ticks

```python
sns.set_style('ticks')
sinplot()
sns.despine()
```

## 3、设置线条大小

```python
# 线条大小
# paper, notebook, talk, poster  从小到大
sns.set_context('poster')
sns.set_style('ticks')
sinplot()
sns.despine()
```
