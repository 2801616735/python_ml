# seaborn单变量分析绘图

```python
import numpy as np
import seaborn as sb
from scipy import stats,integrate
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
sb.set(color_codes=True)
np.random.seed(sum(map(ord,"distributions")))
```

## 1、单特征 直方图

```python
x = np.random.normal(size=100)
sb.distplot(x,kde=False)
```

```python
sb.distplot(x,bins=20,kde=False)
```

## 2、多个特征用散点图

```python
# 生成模拟数据
x,y = [0,1],[(1,0.5),(0.5,1)]
data = np.random.multivariate_normal(x,y,100)
df = pd.DataFrame(data,columns=["x","y"])
df
```

```python
sb.jointplot(x="x",y="y",data=df)
```

```python
with sb.axes_style('white'):
    sb.jointplot(x='x',y='y',kind="hex",color='k',data=df)
```
