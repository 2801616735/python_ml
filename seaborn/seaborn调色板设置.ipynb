## 1、调色板
* color_palette()能传入任何matplotlib所支持的颜色

* set_palette()设置所有图的颜色

## 2、分类色板

```python
# 调色板设置 deep, muted, bright, pastel, dark, colorblind
current_palette = sb.color_palette()
# 展示颜色
sb.palplot(current_palette)
```
